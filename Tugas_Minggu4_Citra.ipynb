```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img_rgb = cv2.imread('doraemon.jpeg')

img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)

c = 255 / (np.log(1 + np.max(img_gray)))
log_image_gray = c * (np.log(img_gray + 1))
log_image_gray = np.array(log_image_gray, dtype=np.uint8)
plt.imshow(log_image_gray, cmap='gray')
plt.title('Citra Logaritmik (dari Grayscale)')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('doraemon.jpeg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
plt.title('Citra Asli (Grayscale)')
plt.show()
```

```python

gamma = 0.5 # Ubah nilai gamma = 0.4, 0.6, 1.2, dst untuk eksperimen
c = 1
power_law = c * np.power(img / 255.0, gamma)
power_law = np.uint8(power_law * 255)
plt.imshow(power_law, cmap='gray')
plt.title(f'Power-Law (Gamma={gamma})')
plt.show()
```

```python

negative = 255 - img
plt.imshow(negative, cmap='gray')
plt.title('Citra Negatif')
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
#Membaca gambar grayscale
img = cv2.imread('gambar.jpeg', cv2.IMREAD_GRAYSCALE)
# Definisi kernel (contoh sharpening)
kernel = np.array([[0, 1, 0],
[-1, 5, -1],
[0, -1, 0]])
# Operasi konvolusi
conv = cv2.filter2D (img, -1, kernel)
# Tampilkan hasil
plt.subplot(1,2,1), plt.imshow(img, cmap='gray'), plt.title('Asli')
plt.subplot(1,2,2), plt.imshow(conv, cmap="gray'), plt.title('Hasil Konvolusi (Sharpen)')
plt.show()
```
